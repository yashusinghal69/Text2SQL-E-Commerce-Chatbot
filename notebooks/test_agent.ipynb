```python
# Define the test_process_question Function

def test_process_question(question: str):
    """
    Test the process_question function and display results.
    
    Args:
        question: Natural language question to test
    """
    print(f"Processing: '{question}'")
    print("=" * 80)
    
    result = process_question(question)
    
    # Display final answer
    print("\n📝 Final Answer:")
    print("-" * 80)
    print(result.get("final_answer", "No answer generated."))
    
    # Display generated SQL query if available
    if result.get("sql_query"):
        print("\n\n🔍 Generated SQL Query:")
        print("-" * 80)
        print(result.get("sql_query"))
    
    # Display graph information
    print(f"\n\n📊 Graph Information:")
    print("-" * 80)
    print(f"Needs Graph: {result.get('needs_graph', False)}")
    if result.get("needs_graph"):
        print(f"Graph Type: {result.get('graph_type', 'N/A')}")
    
    # Display visualization if available
    if result.get("graph_json"):
        print("\n📈 Displaying Visualization...")
        try:
            fig = pio.from_json(result['graph_json'])
            fig.show()
        except Exception as e:
            print(f"Error displaying graph: {e}")
    else:
        if result.get("needs_graph"):
            print("⚠️ Graph was expected but not generated.")
        else:
            print("ℹ️ No graph was needed for this question.")
    
    print("\n" + "=" * 80)
```

```python
# Define the process_question Function

def process_question(question: str) -> dict:
    """
    Process a natural language question and return the final result.
    This is a simple synchronous function for notebook usage.
    
    Args:
        question: Natural language question about the e-commerce data
        
    Returns:
        dict: Final state with answer, SQL query, and graph data if applicable
    """
    initial_state = AgentState(
        question=question,
        sql_query="",
        query_result="",
        final_answer="",
        error="",
        iteration=0,
        needs_graph=False,
        graph_type="",
        graph_json="",
        is_in_scope=True
    )
    
    try:
        # Invoke the graph
        final_state = text2sql_graph.invoke(
            initial_state,
            config={"recursion_limit": 5},
        )
        
        return final_state
        
    except Exception as e:
        return {
            "error": str(e),
            "final_answer": f"An error occurred while processing your question: {str(e)}"
        }
```
